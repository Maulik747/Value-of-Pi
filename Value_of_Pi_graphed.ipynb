```python
import random
import numpy as np
import matplotlib.pyplot as plt
r=10000  # increase the value of r for more accuracy
x=[random.uniform(0,1) for i in range(r)]
y=[random.uniform(0,1) for i in range (r)]

dist=[np.sqrt((x[i]*x[i])+(y[i]*y[i])) for i in range(r)]
a=[]
colmap=np.array(['r','g'])
for i in range(r):
    if dist[i]<1:
        a.append(0)
    else:
        a.append(1)

plt.scatter(x,y,s=1,c=colmap[a])
plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.show()
cnt=0
for j in range (r):
    if a[j]==0:
        cnt=cnt+1

ans= (cnt/r)*(1/(0.5*0.5))
print("The value of pi is: ",ans)
```
